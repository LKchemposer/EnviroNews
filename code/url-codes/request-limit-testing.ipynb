# Testing Request Limit

Because we want to extract full-text articles from various news sources, it is critical that their sites allow us to send and receive requests instantaneously. If not, we may be need to add in some `time.sleep` lines to avoid sending too many requests at a time or some sites are forbidden altogether. Here, I test this issue by sending 50 consecutive requests to various news sites in the newspaper API's `popular_urls`, and getting the status code.

```python
import newspaper
import requests
import pandas as pd
```

```python
checkurls = newspaper.popular_urls()
```

```python
def check_status_code(urls):
    d = dict()
    for url in urls:
        try:
            for _ in range(50):
                res = requests.get(url)
            d[url] = res.status_code
            print(url, res.status_code)
        except:
            continue
    return d
```

```python
stats = check_status_code(checkurls)
```

```python
sites = pd.DataFrame(stats, index=[0]).transpose().reset_index()
sites.columns = ['news_site', 'status_code']
```

```python
def status(row):
    s = row['status_code']
    sc = {404: 'Not Found', 403: 'Forbidden', 429: 'Too Many Requests', 200: 'OK'}
    for i in sc:
        if s == i:
            return sc[i]

sites['status'] = sites.apply(status, axis=1)
```
